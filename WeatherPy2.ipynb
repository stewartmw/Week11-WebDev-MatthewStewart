### Analysis
* To no one's surprise (hopefully), temperature does indeed increase as one approaches the equator.
* Overall, there does not appear to be a strong relationship between humidity and latitude.  However, there does appear to be a cluster of cities between latitudes of -20 and 20 that have 100% humidity.  This makes sense given that the world's tropical rainforests lie between the Tropic of Capricorn (23.5 degrees south) and the Tropic of Cancer (23.5 degrees north).
* There do not appear to be strong relationships between latitude and either wind speed or cloud cover.

### Initial Setup

```python
# Dependencies
import time
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
import warnings
```

```python
# matplotlib global parameters
mpl.rcParams.update({
    'figure.figsize'      : (15,10),
    'font.size'           : 20.0,
    'axes.titlesize'      : 'large',
    'axes.labelsize'      : 'medium',
    'xtick.labelsize'     : 'medium',
    'ytick.labelsize'     : 'medium',
    'legend.fontsize'     : 'large',
    'legend.loc'          : 'upper right'
})
```

```python
# Ignore warnings
warnings.filterwarnings('ignore')
```

### Data Work

```python
# Convert provided cities.csv to dataframe
input_file = 'Resources/cities.csv'

cities_df = pd.read_csv(input_file)
cities_df.head()
```

```python
# dataframe cleanup
cities_df_formatted = cities_df[[
    'City_ID',
    'City',
    'Country',
    'Lat',
    'Lng',
    'Date',
    'Max Temp',
    'Humidity',
    'Wind Speed',
    'Cloudiness'
]]

cities_df_formatted.set_index('City_ID', inplace = True)

cities_df_formatted.rename(columns = {
    'Date': 'Date/Time',
    'Max Temp': 'Max Temp (F)',
    'Humidity': 'Humidity (%)',
    'Wind Speed': 'Wind Speed (mph)',
    'Cloudiness': 'Cloud Cover (%)'
}, inplace = True)

cities_df_formatted['Date/Time'] = cities_df_formatted['Date/Time'].apply(lambda x: time.ctime(x))

cities_df_formatted.head()
```

```python
# Output to new csv
cities_df_formatted.to_csv('Output/cities_formatted.csv')
```

---

### Visualization

```python
# x_axis is always latitude
x_axis = cities_df_formatted['Lat']

# plot temperature vs. latitude
plt.scatter(x = x_axis, y = cities_df_formatted['Max Temp (F)'], alpha = 0.5)
plt.title('Temperature vs. Latitude (6/13/2018)', fontweight = 'bold')
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.savefig('Output/Temp.png')
plt.show()
```

```python
# plot humidity vs. latitude
plt.scatter(x = x_axis, y = cities_df_formatted['Humidity (%)'], alpha = 0.5)
plt.title('Humidity vs. Latitude (6/13/2018)', fontweight = 'bold')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.savefig('Output/Humidity.png')
plt.show()
```

```python
# plot wind speed vs. latitude
plt.scatter(x = x_axis, y = cities_df_formatted['Wind Speed (mph)'], alpha = 0.5)
plt.title('Wind Speed vs. Latitude (6/13/2018)', fontweight = 'bold')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.savefig('Output/Wind.png')
plt.show()
```

```python
# plot cloud cover vs. latitude
plt.scatter(x = x_axis, y = cities_df_formatted['Cloud Cover (%)'], alpha = 0.5)
plt.title('Cloud Cover vs. Latitude (6/13/2018)', fontweight = 'bold')
plt.xlabel('Latitude')
plt.ylabel('Cloud Cover (%)')
plt.savefig('Output/Clouds.png')
plt.show()
```
